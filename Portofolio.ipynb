```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
plt.style.use('fivethirtyeight')
```

```python
data = pd.read_csv('train.csv')
data.head(10)
```

```python
data.info()
```

```python
(data.isna().sum() * 100) / len(data)
```

```python
#check value of median and mean in variable 'Age'

age_med= data['Age'].median()
age_mean = data['Age'].mean()

print(f'Median from "Age": {age_med}')
print(f'Average from "Age": {age_mean}')
```

```python
#fill missing value in 'Age' with median
data['Age'] = data['Age'].fillna(data['Age'].median())
data.info()
```

```python
#drop all Cabin columns
data = data.drop(columns=['Cabin'])
```

```python
#drop missing value in Embarked variable
data = data.dropna()
data.info()
```

```python
#duplicated check
data.duplicated().sum()
```

# EDA

```python
#mencari deskripsi pada kolom Dtype kategordat
catg = ['Sex', 'Ticket', 'Embarked']
data[catg].describe()
```

```python
#1- Survived comparison ratio
f,ax=plt.subplots(1,2,figsize=(18,8))
data['Survived'].value_counts().plot.pie(explode=[0,0.1],autopct='%1.1f%%', ax=ax[0])
ax[0].set_title('Survived')
ax[0].set_ylabel('')
sns.countplot('Survived', data = data, ax=ax[1])
ax[1].set_title('Survived')
plt.show()
```

```python
#2- Comparison of survivors by gender
plt.figure(figsize = (15, 8))
sns.countplot(data = data, x = 'Survived', hue = 'Sex')
plt.show()
```

```python
#3-number of survivors by age
survived = data[data['Survived'].isin([1])]
plt.figure(figsize=(15,8))
sns.histplot(survived['Age'], color = '#05595B')
plt.show()
```

```python
#4-know the city of departure with the number of passengers who survived
plt.figure(figsize=(15,8))
sns.countplot(data = survived, x = 'Embarked', color = '#874356')
plt.show()
```

```python
#5-total number of 'fare' of each city
penumpang = (data
            .groupby(['Embarked', 'Pclass'])
            .agg(total_spend = ('Fare', 'sum'))
            .reset_index())
penumpang
```

```python
plt.figure(figsize = (15, 8))
sns.barplot(data = penumpang, x = 'Embarked', y = 'total_spend', hue = 'Pclass')
plt.show()
```

```python
#6-number of passengers who survived by class 
plt.figure(figsize = (18, 8))
sns.countplot(data = survived, x = 'Pclass', hue = 'Embarked')
plt.show()
```

```python
#correlation
plt.figure(figsize=(15, 8))
correlation = data.corr()
sns.heatmap(correlation, annot = True, fmt = '.2f')
plt.show()

#the highest correlation occurs in Parch with SibSp
```

```python
#pairplot
sns.pairplot(data = data, hue = 'Sex')
```
